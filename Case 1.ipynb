```python
import pandas as pd
import numpy as np
```

```python
df2010 = pd.read_csv('2010.csv')
df2011 = pd.read_csv('2011.csv')
df2012 = pd.read_csv('2012.csv')
df2013 = pd.read_csv('2013.csv')
df2014 = pd.read_csv('2014.csv')
df2015 = pd.read_csv('2015.csv')
df2016 = pd.read_csv('2016.csv')
df2017 = pd.read_csv('2017.csv')
df2018 = pd.read_csv('2018.csv')
df2019 = pd.read_csv('2019.csv')

df = [df2010, df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019]

for i in range(len(df)):
    df[i] = df[i].rename(columns={'Unnamed: 0':'Dates'})
    df[i]['Dates'] = pd.to_datetime(df[i]['Dates'])
    df[i] = df[i].set_index(['Dates'])
```

```python
gasList = []
gasUsage = []
coalList = []
coalUsage = []
plantList = []
plantUsage = []
avg = []
maxVals = []
days = {}
years = list(np.arange(10) + 2010)
for i in range(len(df)):
    gasDays = len(list(filter(lambda x : x == 2, df[i]['AVG_PRICE']))) + 1
    coalDays = len(list(filter(lambda x : x == 3, df[i]['AVG_PRICE']))) + 1
    plantDays = len(df[i]) - gasDays - coalDays
    gasList.append(gasDays)
    gasUsage.append(df[i]['CONSUMPTION'][gasDays])
    coalList.append(coalDays)
    coalUsage.append(df[i]['CONSUMPTION'][coalDays + gasDays] - df[i]['CONSUMPTION'][gasDays])
    plantList.append(plantDays)
    plantUsage.append(df[i]['CONSUMPTION'][-1] - df[i]['CONSUMPTION'][coalDays + gasDays])
    avg.append(df[i]['CONSUMPTION'][-1]/len(df[i]))
    maxVals.append(df[i]['AVG_PRICE'][-1])
    days[years[i]] = (gasDays, coalDays, plantDays)

gasAvg = list(np.array(gasUsage)/np.array(gasList))
coalAvg = list(np.array(coalUsage)/np.array(coalList))
plantAvg = list(np.array(plantUsage)/np.array(plantList))
```

```python
d = {'Years': years, 'Gas Days': gasList, 'Gas Usage': gasUsage, 'Gas Average': gasAvg, 
     'Coal Days': coalList, 'Coal Usage': coalUsage, 'Coal Average': coalAvg, 
     'Plant Days': plantList, 'Plant Usage': plantUsage, 'Plant Average': plantAvg, 
     'Total Average': avg, 'Max Price': maxVals}
df_summary = pd.DataFrame(d)
df_summary = df_summary.append(df_summary.mean(axis=0), ignore_index=True)
```
